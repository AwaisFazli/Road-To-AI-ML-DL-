## Tasks

Task 1: Take input from the user 10 times in the same variable using while loop.

Task 2: Keep asking the user to enter an integer. If the entered integer is an even number, then
break the operation.

Task 3: Ask user to enter a password and save it in a variable pas.
Then ask the user to write again in order to confirm the password (save it in variable pas1).
If the user enters the same password as before then display “Your password is saved” and break
the loop.
If the user enters a different password than before then display “Try again” and start over the
whole process.

Task 4: Save a password in a variable pas.
Ask user to enter password.
If the entered password is same as the password stored in pas, display “Welcome!”
If the user entered wrong password, then display “You entered a wrong password!” and again
start over the process.
If the user enters a wrong password for the 3rd try than display “Too many tries! System is
blocked!” and break the loop.

## Task 1

```python
i = 0
while i < 10:
    a = input("Enter Value ")
    print(a)
    i += 1
```

## Task 2

```python
while True:
    num=int(input("Enter the value :"))
    if num%2==0:
        print("Loop Broke ")
        break
    else:
        print(num," Number Is Odd Keep Entering")
```

## Task 3

```python
while True:
    pas=input("Set The Password :")
    passs=input("Reconfirm The password :")
    if pas==passs:
        print("Password is saved")
        break
    else:
        print("Retry Password")
```

## Task 4

```python
pas="password"
counter=0
while True:
    pas1=input("Enter the password :")
    if pas==pas1:
        print("Welcome!")
        break
    else:
        counter+=1
        if counter==3:
            print("Too many tries, system is blocked !")
            break
        print("You have entered a wrong password!")
```

## Challenge

```python
val = int(input("Enter Value "))
space = val
i = 0

while i < val:
    j = 0
    while j < val:
        print(" ", end = "")
        j += 1
    k = 0    
    while k <= (space - val):
        print("*", end= "")
        k += 1
    print()  
    val -= 1
```
