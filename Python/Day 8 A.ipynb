## Task 1

```python
paisa = int(input("Enter Value: "))

dollars = 0
quarter = 0
dime = 0
nickel = 0
pennies = 0

cur = 0
rem = paisa

print("Amount: ", paisa)
cur = rem // 100
dollars = cur
rem = (rem) - (cur * 100)
print("Dollar:" ,dollars)
        
cur = rem // 25
quarter = cur
rem = rem - (cur * 25)
print("Quarter: ",quarter)
        
cur = rem // 10
dime = cur
rem = rem - (cur * 10)
print("Dime: ",dime)
        
cur = rem // 5
nickel = cur
rem = rem - (cur * 5)
print("Nickel: ",nickel)
        
pennies = rem
print("Pennies: ",pennies )
```

## Task 2

```python

while True:
    
    pounds = int(input("Enter the integer number of pounds (<0 to quit): "))
    if pounds < 0:
        break
    ounces = float(input("Enter the number of ounces as a real number: "))

    if pounds < 0:
        break
    
    pounds1 = ounces / 16
    
    pounds2 = pounds + pounds1
    
    kilo = round(pounds2 / 2.2046, 4)
    print(f"The number of kilograms for {pounds} pounds and {ounces} ounces is {kilo}")
    print()
```
