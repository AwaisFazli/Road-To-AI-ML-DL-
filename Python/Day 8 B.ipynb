## Project

```python
student = []
roll = []
while True:
    print('''
            Choose Option:
            Enter 1 to Add Student
            Enter 2 to List All Student
            Enter 3 Search Student
            Enter 4 Remove Student
            Enter 5 Exit
            ''')
    
    val = int(input("Select Option: "))
    
    if val == 1:
        student.append(input("Enter Name Of Student: "))
        roll.append(input("Enter Roll Number Of Student: "))
        
    elif val == 2:
        print()
        print("List Of Students Are: ")
        for i in student:
            print(i)
        print()
        
    elif val == 3:
        print()
        rol = input("Enter Rol Of Student ")
        
        if rol in roll:
            print(f"Yes Roll# {rol} is Enrolled")
            print()
            
        else:
            print("Data Not Available")
            
    elif val == 4:
        print()
        name = input("Enter Name Of Student")
        rol = input("Enter Rol Of Student ")
        
        if (name in student) and (rol in roll):
            student.remove(name)
            roll.remove(rol)
            
            print(f"Student {name} has been removed")
            print()
        else:
            print("Enter Valid Name Or Roll Number")
    
    elif val == 5:
        print("Shutting Down")
        break
```
