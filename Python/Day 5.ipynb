## Task 1

```python
i = -1
while  i < 10:
    i += 1
    if i == 2 or i == 8:
        continue
    else:
        print(i)
    
```

```python
val = int(input("Enter Value"))

i = 1

while i <= val:
    print("*"*i, end = "")
    i += 1
    print()
    
i = 0
while i <= val :
    print("*"*(val +1), end = "")
    val -= 1
    print()
```
