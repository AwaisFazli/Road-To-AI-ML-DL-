## Task 4

```python
ran = int(input("Enter Range "))
counter = 1
for i in range(1, ran+1):
    if (i == 1) or (i == ran):
        print("- "*(ran))
    else:
        for i in range(1, ran + 1):
            if counter % 2 == 1:
                print("+ ", end = "")
            if counter % 2 == 0:
                print("- ", end = "")
            counter += 1
        print()
                
```
